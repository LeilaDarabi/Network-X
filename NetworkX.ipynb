## Destribution of out_degree for Large graph

```python
out_degrees = dict(g_l_direct.out_degree()) # dictionary node:degree
# filtering nodes outdegree values with outdegree > 0
pos_out_degree_vals = list(filter(lambda val: val > 0, out_degrees.values()))
# getting unique and sorted outdegree values
uq_pos_out_degree_vals = sorted(set(pos_out_degree_vals))
# counting frequency of each outdegree values
out_hist = [pos_out_degree_vals.count(x) for x in uq_pos_out_degree_vals]


import numpy as np
x = np.asarray(uq_pos_out_degree_vals, dtype = float)
y = np.asarray(out_hist, dtype = float)
logx = np.log10(x)
logy = np.log10(y)
plt.figure(figsize=(10,10))
plt.xlim(min(logx), max(logx))
plt.xlabel('log10 (Out Degree)')
plt.ylabel('log10 (Number of nodes)')
plt.title('Out Degree Distribution of Large Graph')
out_degree_dist = plt.plot(logx, logy, 'o')
```

#question 2.6

```python
WCC_m = g_l_direct.subgraph(max(nx.weakly_connected_components (g_m_direct),key=len))
```

```python

frq = {}
for n in WCC_m.nodes :
  ls = dict(nx.single_source_shortest_path_length(WCC_m,n))
  for target , length in ls.items() :
    if length in frq.keys() :
      frq [length] = frq[length] + 1
    else :
      frq [length] = 1
print(frq )
plt.bar(list(frq.keys()),frq.values())
plt.title('Distance Distribution for medium Graph')
plt.ylabel('Frequency')
plt.xlabel('All the Shortest PathLength')
```

```python
WCC_l = g_l_direct.subgraph(max(nx.weakly_connected_components (g_l_direct),key=len))
```

```python

frq = {}
for n in WCC_l.nodes :
  ls = dict(nx.shortest_path_length(WCC_l,n))
  for target , length in ls.items() :
    if length in frq.keys() :
      frq[length] = frq[length] + 1
    else :
      frq[length] = 1
print(frq )
plt.bar( list(frq.keys()),frq.values())
plt.title('Distance Distribution for large graph' )
plt.ylabel('Frequency')
plt.xlabel('All the Shortest PathLength')
```
